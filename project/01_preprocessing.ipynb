### 2. Finding the Area Plygons

```python
df = pd.read_parquet('./data/processed/transactions_clean.parquet')
areas = gpd.read_file('./data/raw/dubai_land_areas.geojson')
```

```python
df['actual_worth_aed'] = df['actual_worth_aed'] / 1000000
df['actual_worth_usd'] = df['actual_worth_usd'] / 1000000
```

#### Approach 1 of merging (names)

```python
areas_names = areas['CNAME_E'].unique()

# Function to get the best match for each area_name
def match_area_name(area_name, choices, threshold=95):
    match, score, _ = process.extractOne(area_name, choices, scorer=fuzz.WRatio)
    if score >= threshold:
        return match
    else:
        return None

df['matched_area'] = df['area_name'].apply(lambda x: match_area_name(x, areas_names))
```

```python
areas = areas[areas['CNAME_E'].isin(df['matched_area'])]
```

```python
areas = areas[['COMM_NUM', 'Name', 'SHAPE.AREA', 'geometry']]

areas['Name'] = areas['Name'].str.lower().str.strip()
areas.rename(columns={'Name':'name', 'COMM_NUM':'comm_number', 'SHAPE.AREA':'area_sqm'}, inplace=True)
```

```python
areas = areas.merge(df[['matched_area', 'area_id']], left_on='name', right_on='matched_area')

areas.drop_duplicates(inplace=True)
areas.reset_index(inplace=True)

areas = areas[['area_id', 'comm_number', 'name', 'area_sqm', 'geometry']]
```

#### Approach 2 of merging (manually created area_id)

```python
areas = areas[areas['area_id'].notna()]

areas = areas[['area_id', 'COMM_NUM', 'Name', 'SHAPE.AREA', 'geometry']]

areas['Name'] = areas['Name'].str.lower().str.strip()
areas.rename(columns={'Name':'name', 'COMM_NUM':'comm_number', 'SHAPE.AREA':'area_sqm'}, inplace=True)
```

```python
areas
```

#### Adding the total worth of the area for each property type (i.e. unit, villa, land, etc.)

```python
def add_property_type_worth(df, areas):
    worth_by_area_type = (
        df.groupby(['area_id', 'property_type'])['actual_worth_usd']
        .sum()
        .reset_index()
)

    worth_pivot = worth_by_area_type.pivot(index='area_id', columns='property_type', values='actual_worth_usd').fillna(0)
    worth_pivot = worth_pivot.reset_index()

    worth_pivot = worth_pivot.rename(columns=lambda x: f'{x.lower()}_total_worth_usd_millions' if x != 'area_id' else x)

    areas_merged = areas.merge(worth_pivot, how='left', on='area_id')

    for col in worth_pivot.columns:
        if col != 'area_id':
            areas_merged[col] = areas_merged[col].fillna(0)
    
    total_worth = (
        df.groupby('area_id')['actual_worth_usd']
        .sum()
        .reset_index()
        .rename(columns={'actual_worth_usd': 'total_worth_usd_millions'})
    )

    areas_merged = areas_merged.merge(total_worth, how='left', on='area_id')
    areas_merged['total_worth_usd_millions'] = areas_merged['total_worth_usd_millions'].fillna(0)

    return areas_merged

def add_property_type_counts(df, areas):
    counts = (
        df.groupby(['area_id', 'property_type'])
        .size()
        .unstack(fill_value=0)
        .reset_index()
    )

    # Merge with areas
    areas_merged = areas.merge(counts, how='left', on='area_id')

    # Fill NaNs for counts
    for col in ['number_of_lands', 'number_of_units', 'number_of_villas', 'number_of_buildings']:
        if col in areas_merged.columns:
            areas_merged[col] = areas_merged[col].fillna(0).astype(int)

    return areas_merged
```

```python
areas = add_property_type_worth(df, areas)
areas = add_property_type_counts(df, areas)
```

```python
areas
```

```python
areas['worth_per_sqm'] = (areas['total_worth_usd_millions'] * 1000000) / (areas['area_sqm'])

areas['unit_worth_per_sqm'] = (areas['unit_total_worth_usd_millions'] * 1000000) / (areas['area_sqm'])
areas['villa_worth_per_sqm'] = (areas['villa_total_worth_usd_millions'] * 1000000) / (areas['area_sqm'])
areas['building_worth_per_sqm'] = (areas['building_total_worth_usd_millions'] * 1000000) / (areas['area_sqm'])
areas['land_worth_per_sqm'] = (areas['land_total_worth_usd_millions'] * 1000000) / (areas['area_sqm'])
```

```python
areas['properties_count'] = areas['unit'] + areas['villa'] + areas['building'] + areas['land']
```

### 4. Preparing The Venues

```python
venues = gpd.read_file('./data/raw/dubai_venues.geojson')
areas = gpd.read_file('./data/processed/transaction_land_areas.geojson')
```

```python
def assign_type(column):
    def detect_place(x):
        if not isinstance(x, str):
            return 'not str'
        x_lower = x.lower().strip()
        if 'restaurant' in x_lower or 'bar' in x_lower or 'cafe' in x_lower or 'café' in x_lower or 'pub' in x_lower:
            return 'restaurant/cafe/bar'
        elif 'food' in x_lower or 'fried' in x_lower or 'sandwich' in x_lower or 'burger' in x_lower or 'bbq' in x_lower or 'shawarma' in x_lower or 'salad' in x_lower or 'hot dog' in x_lower or \
            'pizza' in x_lower or 'bistro' in x_lower or 'snack' in x_lower or 'burrito' in x_lower or 'dinner' in x_lower or 'buffet' in x_lower or 'breakfast' in x_lower:
            return 'fast food'
        elif 'mall' in x_lower:
            return 'mall'
        elif 'shop' in x_lower or 'market' in x_lower or 'store' in x_lower or 'grocery' in x_lower:
            return 'shop/market'
        elif 'gym' in x_lower or 'fitness' in x_lower:
            return 'fitness center'
        elif 'skating' in x_lower or 'golf' in x_lower or 'bowling' in x_lower or 'tennis' in x_lower or 'soccer' in x_lower or 'sports' in x_lower or 'pool' in x_lower or 'cricket' in x_lower or \
            'hockey' in x_lower or 'surf' in x_lower or 'skydiving' in x_lower or 'ski' in x_lower or 'yoga' in x_lower or 'stadium' in x_lower or 'court' in x_lower or 'paintball' in x_lower:
            return 'sports'
        elif 'park' in x_lower or 'entertainment' in x_lower or 'movie' in x_lower or 'art' in x_lower or 'arts' in x_lower or 'museum' in x_lower or 'opera' in x_lower or 'garden' in x_lower or \
            'aquarium' in x_lower or 'zoo' in x_lower or 'harbor' in x_lower or 'concert' in x_lower or 'cultural' in x_lower or 'historic' in x_lower or 'play' in x_lower or 'beach' in x_lower or 'waterfront' in x_lower:
            return 'entertainment'
        else:
            return 'other'
        
    return column.apply(detect_place)
```

```python
venues['venue_type'] = assign_type(venues['Venue Category'])
```

```python
venues.columns = [col.lower() for col in venues.columns]
venues = venues[['comm_num', 'venue id', 'venue', 'rating', 'venue_type', 'geometry']]

venues = venues[venues['venue_type'] != 'other']
```

```python
venues = gpd.GeoDataFrame(venues, geometry='geometry', crs='EPSG:4326')
```

```python
venues = venues.sjoin(areas[['geometry']], predicate='within', how='inner').drop_duplicates('venue id')

venues.drop(columns=['index_right'], inplace=True)
venues.rename(columns={'venue id':'venue_id'}, inplace=True)
```

```python
venues.to_file('./data/processed/dubai_venues_clean.geojson', driver='GeoJSON')
```

### 6. Averagae Distance to Venues

```python
areas = gpd.read_file('./data/processed/transaction_land_areas.geojson')
venues = gpd.read_file('./data/processed/dubai_venues_clean.geojson')
```

```python
distances = []

venue_areas = areas[areas['comm_number'].isin(venues['comm_num'])]

for idx, area in venue_areas.iterrows():
    v = venues[venues['comm_num'] == area['comm_number']]

    distance = pt.single_source_travel_info_keyless(area['geometry'].centroid, v, mode='auto', base_url='https://routing.homenas.us')
    distances.append(distance)

distances = pd.concat(distances, ignore_index=True)
```

```python
distances.drop_duplicates(['comm_num', 'venue_id'], inplace=True)
distances.drop(columns=['geometry'], inplace=True)
```

```python
distances.to_parquet('./data/processed/area_to_venue_distances.parquet')
```

```python
distances_grouped = distances.groupby('comm_num').agg(
    distance_mean=('distance', 'mean'),
    distance_median=('distance', 'median')
).reset_index()
```

```python
distances_grouped
```

```python
areas = areas.merge(distances_grouped, left_on='comm_number', right_on='comm_num', how='left')

areas.drop(columns=['comm_num'], inplace=True)
areas.rename(columns={'distance_mean':'mean_distance_to_venues', 'distance_median':'median_distance_to_venues'}, inplace=True)
```

```python
areas['mean_distance_to_venues'] *= -1
areas['median_distance_to_venues'] *= -1
```

```python
areas.to_file('./data/processed/transaction_land_areas.geojson')
```

### 7. Route Frequencies

```python
venues = gpd.read_file('./data/processed/dubai_venues_clean.geojson')
sample_buildings = gpd.read_file('./data/processed/sample_building_locations.geojson')
```

```python
G = ox.graph_from_place('Dubai')
```

```python
pairs = []
group = venues.groupby('comm_num')

for _, g in group:
    g = g.reset_index(drop=True)

    buildings = sample_buildings[sample_buildings['area_id'] == g['area_id'][0]]

    for _, row in buildings.iterrows():
        for _, dow in g.iterrows():
            pairs.append({
                'source': row['geometry'],
                'destination': dow['geometry']
            })

pairs = pd.DataFrame(pairs)
```

```python
pairs
```

```python
routes = pt.compute_route_frequencies(pairs, source_col='source', dest_col='destination', graph=G)
```

```python
routes.to_file('./data/processed/routes.geojson', driver='GeoJSON')
```
