```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
```

## Part 1: Data Cleaning and Exploration

```python
df = pd.read_csv('./data/crime_data.csv')
df.head()
```

```python
df.isna().sum() # identifying columns that are going to be dropped
```

```python
df_cleaned = df.dropna(axis=1, thresh=(len(df) * 0.5))
df_cleaned
```

```python
df_cleaned.loc[:, "DATE OCC"] = pd.to_datetime(df_cleaned["DATE OCC"], format="%m/%d/%Y %I:%M:%S %p").dt.date

df_cleaned.loc[:, "Year"] = pd.to_datetime(df_cleaned["DATE OCC"]).dt.year
df_cleaned.loc[:, "Month"] = pd.to_datetime(df_cleaned["DATE OCC"]).dt.month
df_cleaned.loc[:, "Day"] = pd.to_datetime(df_cleaned["DATE OCC"]).dt.day

df_cleaned.loc[:, "Hour"] = df_cleaned["TIME OCC"].astype(str).str.zfill(4).str[:2].astype(int)

# Checking the type
type(df_cleaned['DATE OCC'][0])
```

```python
df_filtered = df_cleaned[(df_cleaned['Year'] == 2023) & (df_cleaned['Crm Cd Desc'] == 'BURGLARY')]
df_filtered
```

```python
df_stats = df_cleaned \
    .groupby("AREA NAME") \
    .agg(
        total_crimes=("Vict Age", "count"),
        avg_vict_age=("Vict Age", "mean")
    ) \
    .sort_values(by="total_crimes", ascending=False) 

df_stats
```

## Part 2: Further Exploration (Python only) 

```python
top_crimes = df_cleaned \
    .groupby('Crm Cd Desc') \
    .agg(count=('Crm Cd Desc', 'count')) \
    .sort_values('count', ascending=False) \
    .head(n=3)

top_crimes
```

```python
crime_by_hour = df_cleaned.groupby("Hour").size().reset_index(name="Total_Crimes")
crime_by_hour
```

```python
crime_by_vict_sex = df_cleaned \
    .groupby("Vict Sex") \
    .agg(
        total_crimes=("Vict Age", "count"),
        avg_vict_age=("Vict Age", "mean")
    ) \
    .reset_index()

crime_by_vict_sex
```

## Part 4: Advanced Analysis

```python
df["Severity Score"] = np.where(
    df["Weapon Used Cd"].notna(), 5,
    np.where(df["Crm Cd Desc"] == "BURGLARY", 3, 1)
)

severity_score = df.groupby("AREA NAME")["Severity Score"].sum().reset_index()
severity_score
```
